**문제 10: 이상치 데이터 탐색 및 제거**

데이터프레임에서 점수와 공부시간의 이상치를 처리하세요. 사분위수를 이용한 판별과 제거

```python
#점수에 대해서 사분위수 구하기
Q1_score = df2['점수'].quantile(0.25)
Q3_score = df2['점수'].quantile(0.75)
IQR_score = Q3_score - Q1_score
print(f'Q1_score: {Q1_score}')
print(f'Q3_score: {Q3_score}')
print(f'IQR_score: {IQR_score}')

#공부시간에 대해서 사분위수 구하기
Q1_time = df2['공부시간'].quantile(0.25)
Q3_time = df2['공부시간'].quantile(0.75)
IQR_time = Q3_time - Q1_time
print(f'Q1_time: {Q1_time}')
print(f'Q3_time: {Q3_time}')
print(f'IQR_time: {IQR_time}')
```

```python
#점수의 상한, 하한 계산
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

#공부시간의 상한, 하한 계산
lower_bound_time = Q1_time - 1.5 * IQR_time
upper_bound_time = Q3_time + 1.5 * IQR_time
```

```python
#이상치 처리
outliers_sctm = df2[
    ((df2['점수'] < lower_bound_score) | (df2['점수'] > upper_bound_score)) |
    ((df2['공부시간'] < lower_bound_time) | (df2['공부시간'] > upper_bound_time))
]

outliers_score = df2[(df2['점수'] < lower_bound_score) | (df2['점수'] > upper_bound_score)]
outliers_time = df2[(df2['공부시간'] < lower_bound_time) | (df2['공부시간'] > upper_bound_time)]

print(f'점수 이상치:')
display(outliers_score)
print(f'공부시간 이상치:')
display(outliers_time)

#print(f'이상치 개수: {len(outliers_sctm)}')
#display(outliers_sctm)

#이상치가 제거된 데이터프레임 (이상치 범위 내에 있는 데이터 선택)
cleared_df = df2[
    ((df2['점수'] >= lower_bound_score) & (df2['점수'] <= upper_bound_score)) &
    ((df2['공부시간'] >= lower_bound_time) & (df2['공부시간'] <= upper_bound_time))
]
print(f'이상치가 제거된 데이터프레임:')
#print(f'모든 이상치가 제거된 데이터프레임 크기: {cleared_df.shape}')
#print(f'제거된 이상치 행 수: {len(df2) - len(cleared_df)}')
display(cleared_df)
```

**실습문제**

다음 데이터에 대해 IQR을 계산하고 이상치를 찾아보세요.

```python
data = pd.Series([10, 11, 12, 13, 14, 15, 100, 101])
```

```python
#1. Q1, Q3, IQR 계산
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(f'Q1: {Q1}')
print(f'Q3: {Q3}')
print(f'IQR: {IQR}')
```

```python
#2. 하한, 상한 계산
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f'하한: {lower_bound}')
print(f'상한: {upper_bound}')
```

```python
#3. 이상치 확인
outliers = data[(data < lower_bound) | (data > upper_bound)]
print(f'이상치: {outliers}')
```
