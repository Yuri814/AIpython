```python
import matplotlib.pyplot as plt

year = [2006, 2009, 2012, 2015, 2018]
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]
jpn = [523, 529, 536, 532, 527]

# 그래프가 가로와 세로 크기를 각각 5인치와 3인치 크기 설정
plt.figure(figsize=(5,3))
# year 데이터를 x축으로, kor 데이터를 y축으로 하여 빨간색 점선으로 선 그래프를 그리고 범례에 'Korea'라고 표시
plt.plot(year, kor, 'r--', label='Korea')
plt.plot(year, usa, 'b:', label='Usa' ) #범례에 'Korea'라고 표시
plt.plot(year, jpn, 'g-', label='Japan')
plt.ylim(400, 600)  # y축 범위 설정
# 그래프에 범례를 추가
plt.legend()
# 그래프 제목을 'Mathematics performance'로 설정합니다.
plt.title('Mathematics performance')
plt.xlabel('year')
plt.ylabel('score')
plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2000, 2005, 2010, 2015, 2020]
single_person = [15.5, 20.0, 23.9, 27.2, 31.7]
baby_born = [640089, 438707, 470171, 438420, 272337]

plt.figure(figsize=(5,3))
 # 1인 가구 비율(single_person)과 출생아 수(baby_born) 사이의 관계를 나타내는 산점도를 그립니다.
for i in range(len(year)):
  plt.annotate(year[i],xy=(single_person[i], baby_born[i]),xytext=(5,-10),textcoords='offset points') #각 연도에 해당하는 데이터 지점 옆에 해당 연도를 표시하는 주석을 추가
plt.xlim(14,35)
plt.ylim(230000,700000)
plt.xlabel('single-person household ratio')
plt.ylabel('the number of babies born')
plt.show()
```

```python
import matplotlib.pyplot as plt

exercise ={'climbing':0,'soccer':0,'swimming':0,'tennis':0}

print('좋아하는 운동을 입력하세요.')
print('복수 개의 운동을 입력하려면 콤마로 구분하세요.')

while True:
  s = input('climbing,soccer,swimming,tennis,종료: ')
  if s=='종료':
    break
  for a in s.split(','):
    if a in exercise:
      # 해당 운동의 카운트를 1 증가시키는 연산
    else:
      print(f'{a}는 등록되지 않은 운동입니다.')

plt.figure(figsize=(5.5,3))
plt.bar(exercise.keys(), exercise.values(), color='g', width=0.6)
#plt.xticks(rotation='vertical')
plt.xlabel('exercise')
plt.ylabel('frequency')
plt.show()
```

```python
#8. 연습 문제

#1. 다음 데이터를 이용하여 1년 동안의 월별 평균 강수량을 꺾은선 그래프로 그려보세요.
#월: ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
#강수량(mm): [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]
#그래프 제목과 x, y축 레이블을 추가하세요.

month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rain = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize=(12,6))
sns.lineplot(x=month, y=rain, marker='o', color='g', linestyle='--', label='monthly rainfall')
#sns.lineplot(x=month, y=rain, 'g,o,--') #축약도 가능. 그러나 color, marker, linestyle 순서대로 입력해야함
for i in range(len(rain)):
  plt.annotate(rain[i], xy=(month[i], rain[i]), xytext=(5,-10), textcoords='offset points')
plt.title('Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.legend()
plt.show()

#데이터 포인트에 값 표시
#for i.value in enumerate(rainfall):
#  plt.text(i, value+10, str(value), ha='center', va='bottom') -> value+10 은 값을 annotate할 때 y축 위치, 즉 값의 위 또는 아래
#                                                              -> str 사용 이유 = value 값을 문자열로 바꿔줌
#                                                              -> ha = horizental alignment 왼쪽, 가운데, 오른쪽 정렬 등
#                                                              -> va = vertical alignment 위, 중간, 아래 정렬
```
